```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
ds = load_breast_cancer()
ds.keys()
```

```python
print(ds.DESCR)
```

```python
df = pd.DataFrame(ds.data, columns=ds.feature_names)
df
```

```python
# import klasy związanej z PCA
from sklearn.decomposition import PCA
```

```python
N_DIM = 5
pca = PCA(n_components=N_DIM, random_state=0)
X_pca = pca.fit_transform(df)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
%%timeit -r 10 -n 100
# estymator na danych oryginalnych

X_train, X_test, y_train, y_test = train_test_split(df, ds.target, test_size=.2, random_state=0)
model = DecisionTreeClassifier(max_depth=7, random_state=0)
model.fit(X_train, y_train)
#print(model.score(X_test, y_test))
#pd.DataFrame( confusion_matrix( y_test, model.predict(X_test) ) )
```

```python
%%timeit -r 10 -n 100
# estymator na danych PCA

X_train, X_test, y_train, y_test = train_test_split(X_pca, ds.target, test_size=.2, random_state=0)
model = DecisionTreeClassifier(max_depth=7, random_state=0)
model.fit(X_train, y_train)
#print(model.score(X_test, y_test))
#pd.DataFrame( confusion_matrix( y_test, model.predict(X_test) ) )
```

```python
plt.figure(figsize=(18,18))
df1 = pd.concat([df, pd.Series(ds.target)], axis=1)
sns.heatmap( df1.corr(), annot=True, cmap="jet")
```

```python
best_cols = df1.corr().iloc[-1].abs().sort_values(ascending=False)[1:6].index
best_cols
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[best_cols], ds.target, 
                                                    test_size=.2, random_state=0)
model = DecisionTreeClassifier(max_depth=7, random_state=0)
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
pd.DataFrame( confusion_matrix( y_test, model.predict(X_test) ) )
```
