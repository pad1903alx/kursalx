```python
# konwersja pounds -> kg
df.Weight /= 2.2
```

```python
def convert_height(s):
    # 6-2
    feet, inches = s.split("-")
    return int(feet)*30.4 + int(inches)*2.54

#convert_height("6-10")

df.Height = df.Height.apply(convert_height)
```
