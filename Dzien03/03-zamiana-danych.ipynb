```python
def only_digits(s):
    result = ""
    for ch in s:
        if ch>='0' and ch<='9':
            result += ch
    return result

only_digits("10 packages")
```

```python
%timeit -r 10 -n 1000 df.No_of_pkg.apply(only_digits)
```
