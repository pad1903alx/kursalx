```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4*np.pi, 0.2)
x
```

```python
y = np.sin(x)
y
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, c="#FF0000", linestyle="-.", marker="D") # wyswietla wykres liniowy
plt.title("Wykres nr 1")
plt.xlabel("X")
plt.grid(axis="x")
plt.ylabel("sin(x)")
plt.xticks( np.arange(0, 4.1*np.pi, np.pi) , ['0','Π','2Π','3Π','4Π'] )
plt.yticks([-1,0,1])
plt.show()
```

```python
# wykres punktowy
# (1,1) (2,5) (3,8) (4,10)
plt.plot([1,2,3,4], [1,5,8,10], "r^")
plt.axis([0,5,0,15]);
```

```python
x = np.arange(0,21,1)
plt.figure(figsize=(14,7))
plt.plot(x, x, "g--")
plt.plot(x, 1/2*x**2, "k.-")
plt.plot(x, 1/10*x**3, "r")
plt.text(5, 400, "Teskt na wykresie", fontsize=18, color="magenta")
plt.annotate("Adnotacja na wykresie", [10,100], [15,600], arrowprops={
    "linewidth" : 1, "arrowstyle" : "<->"
}  )
plt.show()
```

### Wykres słupkowy

```python
countries = ["USA","Niemcy","Kanada"]
m_bronze = np.array([12, 10, 8])
m_silver = np.array([4, 2, 8])
m_gold = np.array([2,2,1])

#plt.figure(figsize=(7,5))
plt.bar(countries, m_bronze, color="brown", width=0.5, label="B")
plt.bar(countries, m_silver, color="silver", width=.5, bottom=m_bronze, label="S")
plt.bar(countries, m_gold, color="gold", width=.5, bottom=m_bronze+m_silver, label="G")
plt.legend(loc="upper left", ncol=3)
plt.gcf().set_size_inches(10,5)

plt.yticks(range(0,20,2))

plt.savefig("wykres.png", dpi=300)
plt.savefig("wykres.jpg", quality=30)
plt.savefig("wykres.tiff")

plt.show()
```

```python
countries = ["USA","Niemcy","Kanada"]
x = np.array([1, 5, 9])
m_bronze = np.array([12, 10, 8])
m_silver = np.array([4, 2, 8])
m_gold = np.array([2,2,1])

#plt.figure(figsize=(7,5))
plt.bar(x, m_bronze, color="brown", width=0.5, label="B")
plt.bar(x+0.5, m_silver, color="silver", width=.5,  label="S")
plt.bar(x+1, m_gold, color="gold", width=.5, label="G")
plt.legend(loc="upper left", ncol=3)
plt.gcf().set_size_inches(10,5)

plt.yticks(range(0,20,2))
plt.xticks(x+.5 , countries)


plt.show()
```

### Wykres kołowy

```python
names = ["Jaś", "Małgosia", "Zuzia", "Brajan"]
toys = [10, 7, 8, 13]
colors = ["blue", "yellow", "#FF0033", "lightskyblue"]

plt.figure(figsize=(7,5))
plt.pie(toys, labels=names, autopct="%.1f %%", textprops={
    "fontsize" : 10 , "style":"italic"
}, colors=colors, shadow=True, explode=[0.1, 0, 0.2, 0] )

plt.show()
```
