```python
import pandas as pd
```

### Append

```python
df1 = pd.DataFrame({
   "a"  : [1,2,3,4] , "b":[5,6,7,8], "c":[9,10,11,12], "d":[13,14,15,16]
})
df1
```

```python
df2 = pd.DataFrame({
   "b"  : [1,2,3,4] , "c":[5,6,7,8], "d":[9,10,11,12], "e":[13,14,15,16]
})
df2 *= -1
df2
```

```python
df1.append(df2, ignore_index=True, verify_integrity=True)
```

### Concat

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```

```python
df1.index = [2,3,4,5]
df1
```

```python
pd.concat([df1, df2], axis=1, join='inner')
```
