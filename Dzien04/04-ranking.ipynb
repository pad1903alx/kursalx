```python
import pandas as pd
```

```python
df =pd.read_csv("http://51.91.120.89/extras/rank.csv", usecols=range(1,4))
df
```

```python
df["ranking"] = df.Score.rank(ascending=False)
df.sort_values("ranking")
```

```python
df["ranking"] = df.Score.rank(ascending=False, method="min")
df.sort_values("ranking")
```

```python
df["ranking"] = df.Score.rank(ascending=False, method="max")
df.sort_values("ranking")
```

```python
df["ranking"] = df.Score.rank(ascending=False, method="dense")
df.sort_values("ranking")
```

```python
df["ranking"] = df.Score.rank(ascending=False, method="first")
df.sort_values("ranking")
```
