```python
df.reset_index(inplace=True)
df
```

```python
#dir(df.Data.dt)
df["Year"] = df.Data.dt.year
```

```python
df.groupby("Year")["Zamkniecie"].mean()
```

```python
df["Quarter"] = df.Data.dt.quarter.astype('object')
```

```python
df.groupby(["Year","Quarter"])["Zamkniecie"].mean()
```

```python
# ustaw na nowo Data w indeksie
df.set_index('Data', inplace=True)
```

```python
df.Zamkniecie.resample("AS-MAY").mean()
```

```python
df.Zamkniecie.resample("2W-MON").mean()
```

```python
df.Zamkniecie.resample("14D").max().plot()
```

```python
df.Zamkniecie.resample("21D").min()
```

```python
df.Zamkniecie.resample("21D").max()
```

```python
df.Zamkniecie.resample("21D").agg([min, max]).plot(kind='barh', figsize=(7,19))
```

```python
df.Zamkniecie.plot( color="red", legend=True)

line2 = df.Zamkniecie.rolling(45).mean()
line2.name = "MA45"
line2.plot(figsize=(16,6), grid=True, color="green", legend=True)

line3 = df.Zamkniecie.rolling(90).mean()
line3.name = "MA90"
line3.plot(figsize=(16,6), grid=True, color="yellow", legend=True)
```

### Wstęga Bollingera

```python
WINDOW_SIZE = 40
RATIO = 2.5
df["LowerRibbon"]  = df.Zamkniecie.rolling(WINDOW_SIZE).mean() - RATIO*df.Zamkniecie.rolling(WINDOW_SIZE).std()
df["UpperRibbon"]  = df.Zamkniecie.rolling(WINDOW_SIZE).mean() + RATIO*df.Zamkniecie.rolling(WINDOW_SIZE).std()
df
```

```python
import matplotlib.pyplot as plt
```

```python
df3 = df.loc[ "2021" , ["Zamkniecie","LowerRibbon","UpperRibbon"] ]
df3.plot(figsize=(16,12), grid=True)
plt.fill_between( df3.index, df3.LowerRibbon, df3.UpperRibbon, color="red", alpha=0.1)
```
