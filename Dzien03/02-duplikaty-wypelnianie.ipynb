### Wypełnianie NaN

```python
df = pd.read_csv("http://51.91.120.89/extras/season.csv", na_values=["Not available"])
df
```

```python
df.isna().sum()
```

```python
df.fillna(0)
```

```python
df
```

```python
df.temp.fillna( df.temp.mean() )
```

```python
df.day.fillna(method="ffill", limit=1)
```

```python
df.temp.fillna(method='bfill')
```

```python
df.temp
```

```python
df.temp.interpolate()
```

```python
# dla day - ffill
# dla temp i wind-speed - usrednianie
# dates - konwersja na datetime64

df.day.fillna(method='ffill', inplace=True)
df.temp.fillna( df.temp.mean(), inplace=True )
df["wind-speed"].fillna( df["wind-speed"].mean(), inplace=True)
df.dates = pd.to_datetime(df.dates, format="%d/%m/%Y")
```
