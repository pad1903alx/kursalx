```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colors = ["yellow","red","black","green","blue"]
sns.lmplot(data=df, x="lifeExp", y="gdpPercap", fit_reg=False, aspect=2.5, 
            hue="continent", palette=colors)
plt.grid()
```

```python
url_covid = "https://covid19.who.int/WHO-COVID-19-global-data.csv"
df = pd.read_csv(url_covid)
df.Date_reported = pd.to_datetime(df.Date_reported, format="%Y-%m-%d")
df.sample(n=10, random_state=0)
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=df.query(" Country_code=='PL' and Date_reported>='2021-09-01' "),
            x="Date_reported", y="New_cases", color="red", marker="d")

sns.lineplot(data=df.query(" Country_code=='DE' and Date_reported>='2021-09-01' "),
            x="Date_reported", y="New_cases", color="black", marker="d")

plt.legend(["PL","DE"])
plt.grid()
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=df.query(" Country_code in ['PL','DE','GB'] and Date_reported>='2021-09-01' "),
            x="Date_reported", y="New_cases", hue="Country_code")
```

```python
df.WHO_region.unique()
```

```python
df1 = df.query(" WHO_region=='EURO' and Date_reported>='2021-09-01' and New_cases>=0 and New_deaths>=0 ")
df1.describe()
```

```python
countries = df1.groupby("Country")["New_cases"].sum().sort_values(ascending=False).head(12).index.to_list()
countries
```

```python
plt.subplots(4, 3, figsize=(16, 18))
for index, country in enumerate(countries,1):
    df_tmp = df1.query(f" Country=='{country}' ")
    plt.subplot(4, 3, index)
    plt.plot(df_tmp.Date_reported, df_tmp.New_cases)
    plt.title(country)
    plt.xticks(df_tmp.Date_reported.loc[::30], rotation=40)
plt.tight_layout()
plt.show()
```
