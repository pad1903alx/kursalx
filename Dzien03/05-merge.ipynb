```python
import pandas as pd
```

```python
df = pd.read_csv("http://51.91.120.89/extras/otodom-head-sample.csv", sep=";")
df1 = df[ df.columns[3:] ]
df1.head()
```

```python
df2 = pd.read_csv("http://51.91.120.89/extras/otodom-details-sample.csv", 
                  sep=";", usecols=["ID","descr"])
df2.head()
```

```python
df1.shape, df2.shape
```

```python
# banalne złączenie typu "inner"
pd.merge(df1, df2, how="inner", on="ID")
```

```python
# złączenie na lewą stronę
df3 = pd.merge(df1, df2, how="left", on="ID")
df3
```

```python
df3[ df3.descr.isna() ]
```

```python
# złączenie na lewą stronę
df3 = pd.merge(df1, df2, how="left", on="ID", indicator=True)
df3
```

```python
df3._merge.value_counts()
```

```python
df3.query(" _merge=='left_only' ")
```

```python
df2.rename(columns={"ID":"ogl_ID"}, inplace=True)
df2.columns
```

```python
df3 = pd.merge(df1, df2, left_on="ID", right_on="ogl_ID", how="right", indicator=True)
df3
```

```python
df3._merge.value_counts()
```

```python
dfA = df1.sample(frac=1/3, random_state=0)
dfB = df2.sample(frac=1/3, random_state=0)
```

```python
dfA.shape, dfB.shape
```

```python
df3 = pd.merge(dfA, dfB, left_on="ID", right_on="ogl_ID", how="outer", indicator=True)
df3
```

```python
df3._merge.value_counts()
```

```python
# zachowanie w przypadku duplikacji danych w kolumnie łączącej
df1 = pd.concat([df1, df1.iloc[:3] ] , ignore_index=True) 
df1
```
